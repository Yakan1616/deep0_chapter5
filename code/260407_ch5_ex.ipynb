# 付録 Softmax-with-Loss レイヤの計算グラフ
ソフトマックス関数と交差エントロピー誤差の計算グラフを示し、それらの逆伝播を求めていく.

* ソフトマックス関数：Softmaxレイヤ
* 交差エントロピー誤差：Cross Entropy Errorレイヤ

3クラス分類を行うニューラルネットワークを考える場合

前レイヤからの入力：($a_1$, $a_2$, $a_3$)\
Softmaxレイヤの出力：($y_1$, $y_2$, $y_3$)\
教師ラベル：($t_1$, $t_2$, $t_3$)\
交差エントロピー誤差：$L$

最終的に, Softmax-with-Lossレイヤの逆伝播の結果が($y_1 - t_1$, $y_2 - t_2$, $y_3 - t_3$)となることを示す.

## A.1 順伝播

### Softmaxレイヤの順伝播
ソフトマックス関数は次の数式であらわされる.
$$
y_k = \frac{\exp(a_k)}{\sum_{i=1}^n \exp(a_i)}
\tag{A.1}
$$
これを計算グラフで表すと図A-1のようになる.

図A-2：Softmaxレイヤの計算グラフ\
<img src="A-2.png" width="400"/>

ここでは式(A.1)の分母に当たる$\sum_{i=1}^n \exp(a_i)$を$S$と置く.\

なぜexpを通すのか
* 確率として成立させるため(統計的な対応ができる)
* 単調増加関数であるため(スコアの大小関係を維持できる)
* 勾配計算を行う上で $e$ が便利なため


### 交差エントロピー誤差の順伝播
交差エントロピー誤差は次の数式であらわされる.
$$
L = -\sum_{k} t_k \log y_k
$$
これを計算グラフであらわすと図A-2のようになる.

## A.2 逆伝播

### Cross Entropy Errorレイヤの逆伝播
まずはCross Entropy Errorレイヤの逆伝播を求める.

ポイント
* 逆伝播の最初の値は1
* 「×」ノードは順伝播時の入力をひっくり返して乗算
* 「＋」ノードは順伝播時の入力をそのまま流す
* 「log」ノードの逆伝播は以下の式に従う\
$y = \log x$\
$\frac{\partial y}{\partial x} = \frac{1}{x}
$

入力：1

出力：($-\frac{t_1}{y_1}$, $-\frac{t_2}{y_2}$, $-\frac{t_3}{y_3}$)

### Softmaxレイヤの逆伝播
次にSoftmaxレイヤの逆伝播を求める.

#### ステップ1
前レイヤからの逆伝播の値が流れてくる.\
($-\frac{t_1}{y_1}$, $-\frac{t_2}{y_2}$, $-\frac{t_3}{y_3}$)

#### ステップ2
「×」ノードは順伝播時の入力をひっくり返して乗算する.

($-\frac{t_1}{y_1} \cdot \exp(a_1)$, $-\frac{t_2}{y_2} \cdot \exp(a_2)$, $-\frac{t_3}{y_3} \cdot \exp(a_3)$)

ここで, 

$y_1 = \frac{\exp(a_1)}{S}$ であることから, $\exp(a_1) = y_1 \cdot S$

$\exp(a_k)$を$y_k S$と置き換えると

($-\frac{t_1}{y_1} \cdot y_1 \cdot S$, $-\frac{t_2}{y_2} \cdot y_2 \cdot S$, $-\frac{t_3}{y_3} \cdot y_3 \cdot S$) = ($-t_1 S$, $-t_2 S$, $-t_3 S$)

と書くことができる.

#### ステップ3
「/」ノードの逆伝播は以下の式に従う.\
　$ y = \frac{1}{S} $\
　$ \frac{\partial y}{\partial S} = -\frac{1}{S^2} $

よって,
$-S \cdot (t_1 + t_2 + t_3) \cdot (-\frac{1}{S^2})$ = ($\frac{t_1 + t_2 + t_3}{S}$)

ここで, ($t_1, t_2, t_3$)はone-hotベクトルであるため, $t_1 + t_2 + t_3 = 1$である.

よって, ($\frac{t_1 + t_2 + t_3}{S}$) = ($\frac{1}{S}$)

#### ステップ4
「+」ノードは順伝播時の入力をそのまま流す.

($\frac{1}{s}, \frac{1}{s}, \frac{1}{s}$)

#### ステップ5
「×」ノードは順伝播時の入力をひっくり返して乗算する.

($-\frac{t_1}{y_1} \cdot \frac{1}{S}$, $-\frac{t_2}{y_2} \cdot \frac{1}{S}$, $-\frac{t_3}{y_3} \cdot \frac{1}{S}$)

ここで, 

$y_1 = \frac{\exp(a_1)}{S}$

であることから,

$\frac{1}{S} = \frac{y_1}{\exp(a_1)}$

よって, 

($-\frac{t_1}{\exp(a_1)}$, $-\frac{t_2}{\exp(a_2)}$, $-\frac{t_3}{\exp(a_3)}$)

#### ステップ6
「exp」ノードの逆伝播は以下の式に従う.

　$y = \exp x$\
　$\frac{\partial y}{\partial x} = \exp x$

よって, 

($(\frac{1}{S}-\frac{t_1}{\exp(a_1)}) \cdot \exp(a_1)$, $(\frac{1}{S}-\frac{t_2}{\exp(a_2)}) \cdot \exp(a_2)$, $(\frac{1}{S}-\frac{t_3}{\exp(a_3)}) \cdot \exp(a_3)$)

ここで,

$y_1 = \frac{\exp(a_1)}{S}$

であることから,

$(y_1 - t_1, y_2 - t_2, y_3 - t_3)$
